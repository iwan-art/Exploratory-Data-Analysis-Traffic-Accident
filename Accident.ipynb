```python
import pandas as pd
import numpy as np
```

```python
total = pd.read_excel("ACCIDENT.xlsx", sheet_name = "POLDA")
total
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sns.set()
```

```python
polda = total.reset_index()
plt.figure(figsize = (11,8))
sns.barplot(x = "nama_data",
           y = "value",
           data = polda)
plt.xticks(rotation = 20)
plt.title("Number of Accident based on POLDA")
plt.xlabel("POLDA")
plt.ylabel("Number")


plt.savefig("figure", dpi = 150)
plt.show()
```

dari total 62.975 kasus kecelakaan, daerah Polda Jawa Tengah memiliki tingkat kasus tertinggi dengan 22%

```python
kendaraan = pd.read_excel("ACCIDENT.xlsx", sheet_name = "VEHICLE")
kendaraan
```

```python
kendaraan_viz = pd.crosstab(index = kendaraan["kendaraan"],
                        columns = "Number of Accident",
                        values = kendaraan["value"],
                        aggfunc = "sum").sort_values("Number of Accident", ascending = False)
```

```python
kendaraan_viz
```

```python
kendaraan_viz1 = kendaraan_viz.reset_index()
plt.figure(figsize = (11,8))
sns.barplot(x = "kendaraan",
           y = "Number of Accident",
           data = kendaraan_viz1)
#plt.title("")
plt.xlabel("Vehicle")
plt.ylabel("Number")
plt.savefig("kendaraan", dpi = 150)
plt.show()
```

```python
waktu = pd.read_excel("ACCIDENT.xlsx", sheet_name = "TIME")
waktu
```

```python
plt.figure(figsize = (11,8))
sns.lineplot(data = waktu, 
             x = "waktu", 
             y = "value",
            markers = True)
plt.xlabel("Time of Accident")
plt.ylabel("Number")
plt.savefig("waktu.png", dpi = 150)
```

Jika dilihat berdasarkan waktu, kecelakaan paling sering terjadi di waktu sibuk yaitu sekitar jam 6 AM sampai jam 12 PM
